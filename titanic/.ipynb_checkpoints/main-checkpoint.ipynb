```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn import preprocessing
from sklearn import linear_model
from sklearn import tree
from sklearn.grid_search import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import cross_val_score,StratifiedKFold

import xgboost as xgb

import re
import time
import features
import model
```

```python
data_dir = 'data/Titanic/'
train_df = pd.read_csv(data_dir+'train.csv', header=0)
test_df = pd.read_csv(data_dir+'test.csv', header=0)
all_df=pd.concat([train_df, test_df]).reset_index(drop=True)

all_df.info()
```

```python
all_df.boxplot(column="Fare",by="Survived")
```

```python
all_df.hist(column="Age",by="Survived",bins=30)
```

```python
pclass_surv = pd.crosstab(all_df['Pclass'], all_df['Survived'])
pclass_surv.plot(kind='bar', stacked=True)
```

```python
###########################################################################################################
# generate featres
#
X, y, submit_df, submit_ids, y_weights = features.genDF(all_df, train_df.shape[0])

X_selected, submit_df_selected = features.selectImportance(X, y, submit_df,y_weights)
```

```python
###########################################################################################################
# model LR 
#

lr_X = X_selected.iloc[:,:15]
lr_X_submit = submit_df_selected.iloc[:,:15]

lr_clf = linear_model.LogisticRegression(C=1.0, penalty='l1', tol=1e-6, class_weight={0:1 , 1:  y_weights})
cv = cross_val_score(lr_clf, lr_X.values, y, cv=10)

print 'lr cv:', cv, cv.mean()
```

```python
lr_clf.fit(lr_X.values, y)
lr_pred = lr_clf.predict(lr_X_submit)
print lr_pred.shape
pd.DataFrame({"columns":list(lr_X.columns), "coef":list(lr_clf.coef_.T)})
lr_prob = lr_clf.predict_proba(lr_X_submit)[:,1]
```

```python
###########################################################################################################
# model SVC
#

svc_X = X_selected.iloc[:,:15]
svc_X_submit = submit_df_selected.iloc[:,:15]

from sklearn.svm import SVC
svc_params = {
    'C':0.5,
    'class_weight':{0:1 , 1:  y_weights},
    'random_state' : 7,
    'kernel':'rbf',
    'probability': True,
    'degree': 3
}

svc_clf = SVC(**svc_params)
cv = cross_val_score(svc_clf, svc_X.values ,y, cv=10)
print 'svc cv: ', cv, cv.mean()
```

```python
svc_clf.fit(svc_X.values,y)
svc_prob = svc_clf.predict_proba(svc_X_submit.values)[:,1]
svc_pred = svc_clf.predict(svc_X_submit.values)
svc_pred.shape
```

```python
###########################################################################################################
# model RF
#

rf_columns = filter(lambda x:  re.match(r'perFare|Gender|Fare|Age.*', x) or
                 (re.match(r'.*_scaled|.*_id', x) and  re.match(r'Title|Family|Pclass.*', x)) , X.columns)
rf_X = X[rf_columns]
rf_X_submit = submit_df[rf_columns]
print rf_columns,'\n'

rf_params = { "n_estimators"      : 2000,
                #"max_features"      :  int(np.sqrt(rf_X.shape[1])),
                "min_samples_split" : 0.015,
                "class_weight": {0:1 , 1:  y_weights},
                "max_depth": 7,
                "n_jobs": -1,
                "oob_score": True, 
                "random_state": 7}

rf_clf = RandomForestClassifier(**rf_params)
cv = cross_val_score(rf_clf, rf_X.values, y.values, cv=10)
print 'svc cv: ', cv, cv.mean()
```

```python
rf_clf.fit(rf_X.values, y.values)
rf_prob = rf_clf.predict_proba(rf_X_submit.values)[:,1]
rf_pred = rf_clf.predict(rf_X_submit)
rf_pred.shape
```

```python
###########################################################################################################
# model XGB
#

xgb_X = rf_X
xgb_X_submit = rf_X_submit

xgb_params ={'colsample_bytree':  0.35999999999999999, 
             'scale_pos_weight': 1/y_weights, 
             'learning_rate': 0.005, 
             'nthread': -1,
             'min_child_weight': 9, 
             'n_estimators': 5000, 
             'subsample': 0.7, 
             'reg_lambda': 1.4000000000000001, 
             'seed': 7, 
             'objective': 'binary:logistic', 
             'max_depth': 5,
             'gamma': 0.3}

xgb_clf = xgb.XGBClassifier(**xgb_params)
cv = cross_val_score(xgb_clf, rf_X.values, y.values, cv=10)
print 'xgb cv: ', cv, cv.mean()
```

```python
xgb_clf.fit(xgb_X.values, y.values)
xgb_prob = xgb_clf.predict_proba(xgb_X_submit.values)[:,1]
xgb_pred =  xgb_clf.predict(xgb_X_submit.values)
xgb_pred.shape
```

```python
###########################################################################################################
#  avg predict
#
all_p = np.array(xgb_pred) + np.array(rf_pred) + np.array(lr_pred) + np.array(svc_pred) + np.array(ens_pred) 
avg_p = np.array(map(lambda x: 1 if x>2 else 0, all_p))
np.sum(avg_p ==1)/float(len(avg_p))
```
